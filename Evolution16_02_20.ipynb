```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def allel(_size, maxdev):
#     points=np.linspace(0,10,_size)

    points=np.array(range(_size))
    l=np.zeros(_size)
    res=0
    correction=1.1
    for i in range(1,len(points)):
        dev=np.random.uniform(-1,1)/maxdev
        if l[i-1]>=1:
            l[i]=1
            res=1
        elif l[i-1]<=-1:
            l[i]=-1
            res=-1
        else:
            if dev>0:
                dev=dev*correction
            l[i]=l[i-1]+dev
    return points, l, res
```

```python
pos=[]
for i in range(1000):
    
    num=10
    plus=minus=neutral=0
    for i in range(num):
        f=allel(1000,10)
        if f[2]==1:
            plus+=1
        if f[2]==-1:
            minus+=1
        if f[2]==0:
            neutral+=1
    pos.append(plus)
    
        
#     plt.plot(f[0],f[1])
    
# print (type(plt))
# plt.show()
#     print ('Plus={}, Minus={}, Neutral={}'.format(plus,minus, neutral))
print ('done')
```

```python
def evol(_size, maxdev):
    points=np.array(range(_size))
    l=np.zeros(_size)
    l[0]=0.1
    for i in range(1,len(points)):
        dev=np.random.uniform(-1,1)/maxdev
        if l[i-1]>=1:
            l[i]=1
        elif i>1 and l[i-1]<=0:
            l[i]=0
        else:
            if dev>0:
                l[i]=l[i-1]+dev*10**(-l[i-1])
            else:
                l[i]=l[i-1]+dev
    return points, l
```
